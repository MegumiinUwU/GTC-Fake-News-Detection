*Secondly , we will Preprocess the text using technques such Lowercasing , Tokenization , removing punctuations,numbers,urls,emails , and Lemmatization*

```python
# Text Preprocessing
import spacy, re

nlp = spacy.load("en_core_web_sm")

def preprocessText(text):
    # lowercasing
    text = text.lower()
    # remove mails , numbers , and urls
    text = re.sub(r'http\S+|www\S+|https\S+', '', text)
    text = re.sub(r'\S+@\S+', '', text)
    text = re.sub(r'\d+', '', text)

    # Tokenization
    doc = nlp(text)
    preprocessed = [token for token in doc if not token.is_stop and not token.is_punct and not token.is_space]

    # Lemmatization
    clean = [token.lemma_.strip() for token in preprocessed if token.lemma_.strip()]

    return clean
```

```python
df['Clean Text'] = df['text'].apply(lambda x:preprocessText(x))
```

N-grams Analysis (Top Bigrams/Trigrams)

```python
from collections import Counter
from nltk.util import ngrams
import nltk
nltk.download('punkt')
# Download the specific resource needed for tokenization
nltk.download('punkt_tab')

def get_top_ngrams(corpus, n=None, gram=2):
    vec = []
    for text in corpus:
        tokens = nltk.word_tokenize(str(text).lower())
        vec.extend(list(ngrams(tokens, gram)))
    return Counter(vec).most_common(n)

# Fake news bigrams
print("Top 10 bigrams in Fake News:")
print(get_top_ngrams(df[df['label']==0]['text'], n=10, gram=2))

# Real news bigrams
print("Top 10 bigrams in Real News:")
print(get_top_ngrams(df[df['label']==1]['text'], n=10, gram=2))
```

### **Fourthly, we will perform Feature Engineering to create numerical features from the text. This includes basic text statistics, TF-IDF representations, and sentiment-based features.**

```python
import numpy as np
import re

# text length (number of words)
df['text_length'] = df['text'].apply(lambda x: len(str(x).split()))

# number of unique words
df['unique_words'] = df['text'].apply(lambda x: len(set(str(x).split())))

# punctuation count
df['punct_count'] = df['text'].apply(lambda x: len(re.findall(r'[^\w\s]', str(x))))

# ratio of unique words to total words
df['unique_ratio'] = df['unique_words'] / df['text_length']

print(df[['text_length','unique_words','punct_count','unique_ratio']].head())
```

```python
# Sentiment Features
from textblob import TextBlob

def get_sentiment(text):
    blob = TextBlob(str(text))
    return blob.sentiment.polarity, blob.sentiment.subjectivity

df['sentiment_polarity'], df['sentiment_subjectivity'] = zip(*df['text'].map(get_sentiment))

print(df[['sentiment_polarity','sentiment_subjectivity']].head())
```

```python
# TF-IDF Features

from sklearn.feature_extraction.text import TfidfVectorizer

# word-level + bigram TF-IDF
tfidf_vectorizer = TfidfVectorizer(max_features=5000, ngram_range=(1,2))
tfidf_matrix = tfidf_vectorizer.fit_transform(df['text'])

print("TF-IDF shape:", tfidf_matrix.shape)

# convert to DataFrame
tfidf_df = pd.DataFrame(tfidf_matrix.toarray(), columns=tfidf_vectorizer.get_feature_names_out())
```

```python
# Headline-Text Consistency

from sklearn.metrics.pairwise import cosine_similarity

title_tfidf = tfidf_vectorizer.transform(df['title'].astype(str))
text_tfidf = tfidf_vectorizer.transform(df['text'].astype(str))

cos_sim = []
for i in range(df.shape[0]):
    sim = cosine_similarity(title_tfidf[i], text_tfidf[i])[0][0]
    cos_sim.append(sim)

df['headline_text_similarity'] = cos_sim

print(df[['headline_text_similarity']].head())
```

```python
# Embedding-based Features (Sentence Transformers)
from sentence_transformers import SentenceTransformer
import pandas as pd

# load model
model = SentenceTransformer("all-MiniLM-L6-v2")

# encode texts
embeddings = model.encode(df["text"].astype(str).tolist(),
                          batch_size=64,
                          show_progress_bar=True)

# to DataFrame
embeddings_df = pd.DataFrame(embeddings)
embeddings_df = embeddings_df.add_prefix("embed_")

print("Embeddings shape:", embeddings_df.shape)

# merge with df
df_final = pd.concat([df.reset_index(drop=True), embeddings_df], axis=1)
print("Final dataset shape:", df_final.shape)
```

```python
# Merge All Features
final_df = pd.concat([df.reset_index(drop=True),
                      tfidf_df.reset_index(drop=True),
                      embeddings_df.reset_index(drop=True)], axis=1)

print("Final dataset shape:", final_df.shape)

# Save dataset with engineered features
final_df.to_csv("engineered_fake_news_full.csv", index=False)
print("Feature Engineering completed and saved.")
```
